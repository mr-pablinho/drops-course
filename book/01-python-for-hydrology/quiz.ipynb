```python
q6 = [
    {
        "question": "Which Pandas method changes daily discharge data to monthly mean discharge?",
        "type": "multiple_choice",
        "answers": [
            {"answer": "df.resample('ME').mean()",
             "correct": True,
             "feedback": "Correct! `resample('ME')` groups records by month-end periods; `.mean()` aggregates each group."},
            {"answer": "df.groupby('month').mean()",
             "correct": False,
             "feedback": "This gives the climatological mean (averaged across all years) for each calendar month — not a monthly time series."},
            {"answer": "df.rolling(30).mean()",
             "correct": False,
             "feedback": "Rolling window smooths without reducing the number of rows; it does not aggregate to monthly time steps."},
            {"answer": "df.interpolate(method='monthly')",
             "correct": False,
             "feedback": "`.interpolate()` fills gaps — it does not aggregate."}
        ]
    }
]

display_quiz(q6)
```
